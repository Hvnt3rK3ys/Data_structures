<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: int </h3>

````python
#include <iostream>

int main() {
    int num = 5;

    // Mostramos la dirección de memoria de la variable 'num'
    std::cout << "Dirección de memoria de 'num' antes de la modificación: " << &num << std::endl;

    // Modificamos el valor de 'num'
    num = 10;

    // Mostramos la dirección de memoria de la variable 'num' después de la modificación
    std::cout << "Dirección de memoria de 'num' después de la modificación: " << &num << std::endl;

    return 0;
}
````

```

Dirección de memoria de 'num' antes de la modificación: 0x7ffe71f9a6e4
Dirección de memoria de 'num' después de la modificación: 0x7ffe71f9a6e4

```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: float </h3>

````python
#include <iostream>

int main() {
    float x = 3.14;
    std::cout << "Dirección de memoria de x: " << &x << std::endl;

    // Cambiamos el valor de x
    x = 42.0;
    std::cout << "Nuevo valor de x: " << x << std::endl;
    std::cout << "Dirección de memoria de x: " << &x << std::endl;

    return 0;
}
````

```

Dirección de memoria de x: 0x7ffc9d4913c4
Nuevo valor de x: 42
Dirección de memoria de x: 0x7ffc9d4913c4


```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: bool</h3>

````python
#include <iostream>

int main() {
    bool myBool = true;
    
    // Imprimimos la dirección de memoria de la variable bool original
    std::cout << "Dirección de memoria de myBool antes de modificar: " << &myBool << std::endl;

    // Intentamos modificar la variable bool
    myBool = false;

    // Imprimimos la dirección de memoria de la variable bool después de intentar modificar
    std::cout << "Dirección de memoria de myBool después de modificar: " << &myBool << std::endl;

    return 0;
}
````

```

Dirección de memoria de myBool antes de modificar: 0x7ffdb9f7ea37
Dirección de memoria de myBool después de modificar: 0x7ffdb9f7ea37


```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: char </h3>

````python
#include <iostream>

int main() {
    char letra = 'A';
    
    std::cout << "Valor original de letra: " << letra << std::endl;
    std::cout << "Dirección de memoria original: " << static_cast<void*>(&letra) << std::endl;

    // Intentamos modificar la letra
    letra = 'B';

    std::cout << "Valor modificado de letra: " << letra << std::endl;
    std::cout << "Dirección de memoria después de modificar: " << static_cast<void*>(&letra) << std::endl;

    return 0;
}
````

```

Valor original de letra: A
Dirección de memoria original: 0x7ffdc27fa6a7
Valor modificado de letra: B
Dirección de memoria después de modificar: 0x7ffdc27fa6a7


```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: double </h3>

````python
#include <iostream>

int main() {
    double num = 3.14; // Crear una variable double

    // Imprimir la dirección de memoria de la variable
    std::cout << "Dirección de memoria de num antes de la modificación: " << &num << std::endl;

    // Intentar modificar el valor de num
    num = 42.0;

    // Imprimir la dirección de memoria de la variable después de la modificación
    std::cout << "Dirección de memoria de num después de la modificación: " << &num << std::endl;

    return 0;
}
````

