<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: int </h3>

````python
public class Main {
    public static void main(String[] args) {
        int num1 = 10; // Creamos una variable int y le asignamos el valor 10
        System.out.println("Antes de modificar el valor: " + num1);
        
        modificarValor(num1); // Intentamos modificar el valor
        
        System.out.println("Después de modificar el valor: " + num1);
    }
    
    public static void modificarValor(int num) {
        num = 20; // Intentamos modificar el valor a 20
        System.out.println("Dentro del método: " + num);
    }
}
````

```
Antes de modificar el valor: 10
Dentro del método: 20
Después de modificar el valor: 10

```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: float </h3>

````python
public class Main {
    public static void main(String[] args) {
        float originalFloat = 3.14f;
        System.out.println("Valor original: " + originalFloat);
        
        // Intentamos modificar el valor original
        modificarFloat(originalFloat);
        
        // Verificamos si el valor original cambió
        System.out.println("Valor después de intentar modificar: " + originalFloat);
    }
    
    // Método para intentar modificar el float
    public static void modificarFloat(float f) {
        f = 42.0f; // Modificamos el valor dentro del método
    }
}
````

```
Valor original: 3.14
Valor después de intentar modificar: 3.14

```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: str</h3>

````python
public class Main {
    public static void main(String[] args) {
        String str1 = "Hola";
        System.out.println("Antes de la modificación: str1 = " + str1 + ", hashcode: " + System.identityHashCode(str1));

        str1 = str1 + " Mundo"; // Concatenamos " Mundo" a str1
        System.out.println("Después de la modificación: str1 = " + str1 + ", hashcode: " + System.identityHashCode(str1));
    }
}
````

```
Antes de la modificación: str1 = Hola, hashcode: 1608446010
Después de la modificación: str1 = Hola Mundo, hashcode: 1870252780

```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: bool </h3>

````python
public class Main {
    public static void main(String[] args) {
        boolean original = true;
        System.out.println("Original: " + original);

        // Intentamos modificar el valor original
        original = false;
        System.out.println("Después de modificar: " + original);

        // Verificamos las referencias en memoria
        System.out.println("Referencia original: " + System.identityHashCode(original));

        // Creamos un nuevo booleano con el valor modificado
        boolean nuevo = false;
        System.out.println("Nuevo booleano: " + System.identityHashCode(nuevo));
    }
}
````

```
Original: true
Después de modificar: false
Referencia original: 1338823963
Nuevo booleano: 1338823963

```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: char </h3>

````python
public class Main {
    public static void main(String[] args) {
        // Crear un objeto char
        char c = 'A';
        
        // Mostrar la referencia en memoria del char antes de intentar modificarlo
        System.out.println("Referencia en memoria antes de la modificación: " + System.identityHashCode(c));
        
        // Intentar modificar el char
        c = 'B';
        
        // Mostrar la referencia en memoria del char después de intentar modificarlo
        System.out.println("Referencia en memoria después de la modificación: " + System.identityHashCode(c));
    }
}
````

```
Referencia en memoria antes de la modificación: 992136656
Referencia en memoria después de la modificación: 1156060786

```
