<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: int </h3>

```python
%%time
# Definimos una función para comprobar si un dato es mutable o inmutable
def comprobar_mutabilidad(dato):
    try:
        dato += 1  # Intentamos modificar el valor del dato
        print("El dato es mutable")
    except TypeError:
        print("El dato es inmutable")

# Creamos una variable de tipo int
numero = 5

# Comprobamos si el tipo de dato int es mutable o inmutable
comprobar_mutabilidad(numero)
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: float </h3>

```python
%%time
# Definimos una variable con un valor float
numero = 3.14

# Imprimimos el id del objeto para identificar su ubicación en memoria
print("ID original:", id(numero))

# Intentamos modificar el valor del float
numero += 1.0

# Imprimimos el id del objeto después de intentar modificarlo
print("ID después de modificar:", id(numero))

# Intentamos modificar el valor del float usando el método de asignación
numero = 6.28

# Imprimimos el id del objeto después de intentar modificarlo usando el método de asignación
print("ID después de asignación:", id(numero))
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: str</h3>

```python
%%time
cadena = "Hola mundo"

# Intentamos modificar la cadena
try:
    cadena[0] = 'h'  # Intentamos cambiar la primera letra a minúscula
    print("La cadena se modificó con éxito.")
except TypeError:
    print("La cadena no se puede modificar.")

print("La cadena original es:", cadena)
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: bool </h3>

```python
%%time
# Definimos una función para verificar si un objeto es mutable o inmutable
def verificar_mutabilidad(objeto):
    inicial = objeto  # Guardamos el valor inicial del objeto
    objeto = not objeto  # Intentamos modificar el objeto
    if objeto is inicial:
        return "El tipo de dato es inmutable"
    else:
        return "El tipo de dato es mutable"

# Probamos con un objeto bool inicializado como True
booleano = True
resultado = verificar_mutabilidad(booleano)
print("Para el tipo de dato bool:", resultado)
```
