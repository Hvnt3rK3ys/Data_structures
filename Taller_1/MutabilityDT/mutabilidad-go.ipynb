<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: int </h3>

````python
package main

import "fmt"

func main() {
    // Definimos una variable de tipo int
    var x int = 5

    fmt.Println("Valor original de x:", x)

    // Intentamos modificar el valor de x
    modificarInt(x)

    // Imprimimos el valor de x después de intentar modificarlo
    fmt.Println("Valor de x después de intentar modificarlo:", x)
}

func modificarInt(a int) {
    // Intentamos modificar el valor de a
    a = 10
}
````

```
Valor original de x: 5
Valor de x después de intentar modificarlo: 5

Program exited.
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: float </h3>

````python
package main

import "fmt"

func main() {
    // Definimos una variable de tipo float64
    var x float64 = 3.14

    // Mostramos el valor original
    fmt.Println("Valor original de x:", x)

    // Intentamos cambiar el valor de x
    cambiarValor(x)

    // Mostramos el valor de x después del intento de cambio
    fmt.Println("Valor de x después de intentar cambiarlo:", x)
}

func cambiarValor(valor float64) {
    // Intentamos cambiar el valor de la variable local
    valor = 2.718
    fmt.Println("Valor cambiado dentro de la función:", valor)
}
````

```
Valor original de x: 3.14
Valor cambiado dentro de la función: 2.718
Valor de x después de intentar cambiarlo: 3.14

Program exited.
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: str</h3>

````python
package main

import (
    "fmt"
)

func main() {
    // Definimos una cadena inicial
    cadena := "Hola, mundo"

    // Intentamos modificar la cadena original
    modificarCadena(cadena)

    // Imprimimos la cadena original
    fmt.Println("Cadena original:", cadena)
}

func modificarCadena(s string) {
    // Intentamos modificar la cadena recibida
    // Concatenamos una cadena adicional
    s += ", ¡adiós!"
}
````

```
Cadena original: Hola, mundo

Program exited.
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: bool </h3>

````python
package main

import "fmt"

func main() {
    // Definimos una variable de tipo bool
    var b bool = true
    fmt.Println("Valor original de b:", b)

    // Intentamos cambiar el valor de b
    cambiarValor(b)
    fmt.Println("Valor de b después de la función:", b)
}

func cambiarValor(valor bool) {
    // Intentamos cambiar el valor del parámetro valor
    valor = false
}
````

```
Valor original de b: true
Valor de b después de la función: true

Program exited.
```

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 2rem; text-align: center;padding:0.5rem;border-radius:1rem;"> TD: complex </h3>

````python
package main

import "fmt"

func main() {
    // Definimos un complex inicial
    c := complex(3, 4)
    fmt.Println("Complex inicial:", c)

    // Intentamos modificar el complex
    c = complex(5, 6) // Esto no cambiará el valor de c, en su lugar, se creará un nuevo complex
    fmt.Println("Complex modificado:", c)

    // Mostramos el complex original, que permanece sin cambios
    fmt.Println("Complex original:", complex(3, 4))
}
````

```
Complex inicial: (3+4i)
Complex modificado: (5+6i)
Complex original: (3+4i)

Program exited.
```
