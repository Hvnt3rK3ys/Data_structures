Implementar un algoritmo de multiplicación de matrices en Python que exija capacidad de cálculo y medir su tiempo de ejecución en diferentes entornos (CPU, GPU, TPU).

**Implementación del Algoritmo:**
* Escribir un script en Python para multiplicar dos matrices grandes utilizando Numpy o listas de Python. 
* Asegurarse de que las matrices sean lo suficientemente grandes como para poder observar diferencias significativas en el tiempo de ejecución entre los diferentes entornos.



<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 1rem; text-align: center;padding:0.5rem;border-radius:1rem;"> Resultados en CPU</h3>

````python
%%time
class Producto_M:
    def __init__(self, matriz1, matriz2):
        self.matriz1 = matriz1
        self.matriz2 = matriz2

    def multiplicar_matrices(self):
        # Verificar si las dimensiones son adecuadas para la multiplicación
        if self.matriz1.shape[1] != self.matriz2.shape[0]:
            raise ValueError("Las dimensiones de las matrices no son compatibles para la multiplicación.")

        # Realizar la multiplicación de matrices
        resultado = np.dot(self.matriz1, self.matriz2)

        return resultado
````

````python
%%time
%%pyinstrument


    
# Definir dos matrices grandes
matriz1 = np.random.rand(15000, 15000)
matriz2 = np.random.rand(15000, 15000)


# Crear instancia de Producto_M
producto = Producto_M(matriz1, matriz2)

# Multiplicar las matrices
resultado = producto.multiplicar_matrices()

#print(f"Resultado de la multiplicación de matrices:{resultado}")
print("\n ------------------- TIME PERFORMANCE --------------------- \n\n")

cProfile.run('producto.multiplicar_matrices()')
````

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 1rem; text-align: center;padding:0.5rem;border-radius:1rem;"> Resultados en GPU</h3>

````python
%%time
class Producto_M_GPU:
    def __init__(self, matriz1, matriz2):
        self.matriz1 = cp.asarray(matriz1)
        self.matriz2 = cp.asarray(matriz2)

    def multiplicar_matrices(self):
        # Verificar si las dimensiones son adecuadas para la multiplicación
        if self.matriz1.shape[1] != self.matriz2.shape[0]:
            raise ValueError("Las dimensiones de las matrices no son compatibles para la multiplicación.")

        # Realizar la multiplicación de matrices en la GPU
        resultado = cp.dot(self.matriz1, self.matriz2)

        return resultado
````

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 1rem; text-align: center;padding:0.5rem;border-radius:1rem;"> Resultados en TPU</h3>

````python
%%time

class Producto_M_TPU:
    def __init__(self, matriz1, matriz2):
        self.matriz1 = matriz1
        self.matriz2 = matriz2

    def multiplicar_matrices(self):
        # Convertir las matrices a tensores de TensorFlow
        tensor1 = tf.constant(self.matriz1)
        tensor2 = tf.constant(self.matriz2)
        
        # Verificar si las dimensiones son adecuadas para la multiplicación
        if self.matriz1.shape[1] != self.matriz2.shape[0]:
            raise ValueError("Las dimensiones de las matrices no son compatibles para la multiplicación.")

        # Realizar la multiplicación de matrices en la TPU
        with tf.device('/device:TPU:0'):
            resultado = tf.matmul(tensor1, tensor2)

        return resultado

    
````

````python
%%time
%%pyinstrument

# Definir dos matrices grandes
matriz1 = tf.random.uniform(shape=(15000, 15000))
matriz2 = tf.random.uniform(shape=(15000, 15000))

# Crear instancia de Producto_M_TPU
producto_tpu = Producto_M_TPU(matriz1, matriz2)

# Multiplicar las matrices en la TPU
resultado_tpu = producto_tpu.multiplicar_matrices()


#print(f"Resultado de la multiplicación de matrices:{resultado_tpu}")

print("\n ------------------- TIME PERFORMANCE --------------------- \n\n")

cProfile.run('producto_tpu.multiplicar_matrices()')
````

### Objetivo
Implementar una operación de multiplicación de tensores utilizando TensorFlow y comparar el rendimiento de la ejecución en CPU, GPU y TPU.

**Implementación de la Operación con Tensores:**
* Utilizar TensorFlow para implementar la multiplicación de dos tensores grandes.

<h3 style="color:#0c0a0b;background-color:#f4a261;font-size: 1rem; text-align: center;padding:0.5rem;border-radius:1rem;"> Resultados en CPU</h3>

````python
%%time

class Product_Tns:
    def __init__(self, tensor1, tensor2):
        self.tensor1 = tensor1
        self.tensor2 = tensor2

    def product(self):
        # Verificar si las dimensiones son adecuadas para el producto
        if self.tensor1.shape != self.tensor2.shape:
            raise ValueError("Las dimensiones de los tensores no son iguales para el producto.")

        # Realizar el producto de tensores
        resultado = tf.multiply(self.tensor1, self.tensor2)

        return resultado

    
````

````python
%%time
%%pyinstrument
    
# Definir dos tensores
tensor1 = tf.random.uniform(shape=(15000, 15000))
tensor2 = tf.random.uniform(shape=(15000, 15000))

# Crear instancia de Product_Tns
producto_tns = Product_Tns(tensor1, tensor2)

# Realizar el producto de tensores en CPU
resultado_tns = producto_tns.product()


#print(f"Resultado del producto de tensores en CPU:{resultado_tns}")
print("\n ------------------- TIME PERFORMANCE --------------------- \n\n")

cProfile.run('producto_tns.product()')
````
